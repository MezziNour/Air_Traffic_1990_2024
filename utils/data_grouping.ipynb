### PART 1 : Airport Dataset

Data importation : 

```python
data = pd.read_csv('../data/APT/processed/trafic_aerien_1990_2024.csv', sep=';')
```

Giving meaningful names to my dataset : 

```python
data = data.rename(columns={
    'ANMOIS': 'annee_mois',
    'APT': 'code_aeroport',
    'APT_NOM': 'nom_aeroport',
    'APT_ZON': 'zone',
    'APT_PEQ': 'unites_trafic',
    'APT_PAX_dep': 'passagers_depart',
    'APT_PAX_arr': 'passagers_arrivee',
    'APT_PAX_tr': 'passagers_transit',
    'APT_FRP_dep': 'fret_depart',
    'APT_FRP_arr': 'fret_arrivee',
    'APT_NMVT_mxt': 'mouvements_passagers',
    'APT_NMVT_cgo': 'mouvements_cargo',
    'source_file': 'source'
})
```

```python
data['annee'] = data['annee_mois'].astype(str).str[:4].astype(int)
data['mois'] = data['annee_mois'].astype(str).str[4:6].astype(int)
data.head()
```

```python
import pandas as pd

apt_codes = data["code_aeroport"].unique().tolist()


airports = pd.read_csv("../data/APT/raw/airports.csv")


airports_subset = airports[airports["ident"].isin(apt_codes)].copy()

coords_df = airports_subset[[
    "ident",
    "municipality",
    "latitude_deg",
    "longitude_deg",
]].rename(columns={
    "ident": "code_aeroport",
    "municipality": "ville",
    "latitude_deg": "latitude",
    "longitude_deg": "longitude"
})


data = data.merge(coords_df, on="code_aeroport", how="left")


missing_codes = sorted(set(apt_codes) - set(coords_df["code_aeroport"]))
print(f"{len(coords_df)} aéroports trouvés sur {len(apt_codes)} codes uniques.")
if missing_codes:
    print("Codes non trouvés dans OurAirports (à compléter manuellement) :")
    print(missing_codes)
else:
    print("Tous les codes ont été trouvés !")


display(data.head())
```

```python

manual_airports = pd.DataFrame([
    {"code_aeroport": "LFBV", "ville": "Le Castellet", "latitude": 43.241, "longitude": 5.802},
    {"code_aeroport": "LFSF", "ville": "Saint-Flour", "latitude": 44.923, "longitude": 3.086},
    {"code_aeroport": "LFSR", "ville": "Saint-Raphaël", "latitude": 43.430, "longitude": 6.767},
    {"code_aeroport": "LFTG", "ville": "Hyères", "latitude": 43.097, "longitude": 6.148},
    {"code_aeroport": "ZZMT", "ville": "Inconnu", "latitude": None, "longitude": None},
    {"code_aeroport": "ZZOM", "ville": "Inconnu", "latitude": None, "longitude": None},
])


coords_df = pd.concat([coords_df, manual_airports], ignore_index=True)


data = data.drop(columns=["nom_aeroport_officiel", "ville", "latitude", "longitude", "pays_iso"], errors='ignore')
data = data.merge(coords_df, on="code_aeroport", how="left")


missing_codes = sorted(set(apt_codes) - set(coords_df["code_aeroport"]))
print(f"{len(coords_df)} aéroports trouvés sur {len(apt_codes)} codes uniques.")
if missing_codes:
    print("Codes non trouvés dans OurAirports (à compléter manuellement) :")
    print(missing_codes)
else:
    print("Tous les codes ont été trouvés !")

display(data.head())
```

### PART 2 : Liaison Dataset

```python
folder_path = os.path.join("..","data", "LSN", "raw")
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
dfs = []
for file in sorted(csv_files):
    try:
        df = pd.read_csv(file, sep=";", encoding="utf-8", low_memory=False)
        df["source_file"] = os.path.basename(file)
        dfs.append(df)
    except Exception as e:
        print(f"Error loading {file}: {e}")
```

```python
merged_df = pd.concat(dfs, ignore_index=True)
print(f"\nMerged dataset shape: {merged_df.shape}")
```

```python
merged_df.duplicated().sum()
```

```python
merged_df.isna().sum()
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

```python
merged_df["ANNEE"] = merged_df["ANMOIS"].astype(str).str[:4].astype(int)
merged_df["MOIS"] = merged_df["ANMOIS"].astype(str).str[4:].astype(int)
```

```python
merged_df.to_csv('../data/LSN/processed/lsn.csv', sep=';', index=False)
```
